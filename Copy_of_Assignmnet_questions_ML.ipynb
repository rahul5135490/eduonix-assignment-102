### Top 10 countries with most number of players

```python
top_countries_name=df["Nationality"].value_counts().head(10) # assigning top 10 values by using head() function
top_countries_name
```

```python
# Group data by nationality and count players
country_counts = df['Nationality'].value_counts()

# Select the top 10 countries
top_countries = country_counts.head(10)

# Plot the results using a bar chart
plt.figure(figsize=(12, 6))
top_countries.plot(kind='bar',color="red")
plt.title("Top 10 Countries with Most Players")
plt.xlabel("Countries")
plt.ylabel("Number of Players")
plt.xticks(rotation=90, ha="right")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,10))
x = top_countries_name.values# x axis
y = top_countries_name.index # y axis
ax.barh(y, x, align='center', color='green')
ax.set_xlabel('Number of Players')
ax.set_ylabel('Name of Countries')
ax.set_title('Top 10 Countries with most number of players')
plt.show()
```

### Age distribution in top countries

```python
top_countries_name
```

```python
# Determine the top countries based on the number of players
top_countries = df['Nationality'].value_counts().head(10).index.tolist()

# Filter the dataset to include only players from the top countries
top_countries_data = df[df['Nationality'].isin(top_countries)]

# Plot the age distribution for players from the top countries
plt.figure(figsize=(12, 6))
for country in top_countries:
    country_data = top_countries_data[top_countries_data['Nationality'] == country]
    plt.hist(country_data['Age'], bins=20, alpha=0.6, label=country)

plt.title("Age Distribution of Players in Top Countries")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend()
plt.show()
```

### What is the distribution of overall score of players from top countries?

```python
# Determine the top countries based on the number of players
top_countries = df['Nationality'].value_counts().head(10).index.tolist()

# Filter the dataset to include only players from the top countries
top_countries_data = df[df['Nationality'].isin(top_countries)]

# Plot the distribution of overall scores for players from the top countries
plt.figure(figsize=(12, 6))
for country in top_countries:
    country_data = top_countries_data[top_countries_data['Nationality'] == country]
    plt.hist(country_data['Overall'], bins=20, alpha=0.6, label=country)

plt.title("Overall Score Distribution of Players in Top Countries")
plt.xlabel("Overall Score")
plt.ylabel("Count")
plt.legend()
plt.show()
```
