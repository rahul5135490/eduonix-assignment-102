# 8.Calculate IQR(difference between 75% and 25% quartile)

```python
# PRINT IQR OF MONTHLY HH INCOME
```

```python
iqr_Mthly_HH_Income = df['Mthly_HH_Income'][df['Mthly_HH_Income'].between(df['Mthly_HH_Income'].quantile(.25), df['Mthly_HH_Income'].quantile(.75), inclusive=True)]
```

```python
print(iqr_Mthly_HH_Income)
```

```python
# PRINT IQR OF MONTHLY HH INCOME
Q1 = df.Mthly_HH_Income.quantile(0.25)
Q3 = df.Mthly_HH_Income.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# PRINT IQR OF MONTHLY HH EXPENSE
Q1 = df.Mthly_HH_Expense.quantile(0.25)
Q3 = df.Mthly_HH_Expense.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# PRINT IQR FOR NO. OF FAMILY MEMBERS
Q1 = df.No_of_Fly_Members.quantile(0.25)
Q3 = df.No_of_Fly_Members.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# PRINT IQR OF EMI OR HOUSE RENTAL 
Q1 = df.Emi_or_Rent_Amt.quantile(0.25)
Q3 = df.Emi_or_Rent_Amt.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# PRINT IQR OF ANNUAL HOUSEHOLD INCOME
Q1 = df.Annual_HH_Income.quantile(0.25)
Q3 = df.Annual_HH_Income.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# PRINT IQR OF NUMBER OF EARNING MEMBERS
Q1 = df.No_of_Earning_Members.quantile(0.25)
Q3 = df.No_of_Earning_Members.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```
